```python
#Fitting the graph

import numpy as np
import matplotlib.pyplot as plt

N = 100

#Using np.random.randint we can create random points for our x and y values
x = np.random.randint(1000,size=N)
y = np.random.randint(1000,size=N)

#Creates a range for the x axis for fitting functions and graphing
xdim = np.linspace(0,1000,N)

#Linear fit
fit1 = np.polyfit(xdim,y,1,full=True)
y_p1 = np.polyval(fit1[0],xdim)

#Quadratic fit
fit2 = np.polyfit(xdim,y,2,full=True)
y_p2 = np.polyval(fit2[0],xdim)

#Cubic fit
fit3 = np.polyfit(xdim,y,3,full=True)
y_p3 = np.polyval(fit3[0],xdim)

#4th degree fitting
fit4 = np.polyfit(xdim,y,4,full=True)
y_p4 = np.polyval(fit4[0],xdim)

#5th degree fitting
fit5 = np.polyfit(xdim,y,5,full=True)
y_p5 = np.polyval(fit5[0],xdim)

#6th degree fitting
fit6 = np.polyfit(xdim,y,6,full=True)
y_p6 = np.polyval(fit6[0],xdim)

#Plots for random data and the fits, all on one graph
plt.plot(x,y,'.',label = 'Random Noise')
plt.plot(xdim,y_p1,label = '1st degree')
plt.plot(xdim,y_p2,label = '2nd degree')
plt.plot(xdim,y_p3,label = '3rd degree')
plt.plot(xdim,y_p4,label = '4th degree')
plt.plot(xdim,y_p5,label = '5th degree')
plt.plot(xdim,y_p6,label = '6th degree')
plt.legend()
plt.show()

#Residual of each fit to see how accurate it is
print("The residual for 1st degree fitting is", fit1[1])
print("The residual for 2nd degree fitting is", fit2[1])
print("The residual for 3rd degree fitting is", fit3[1])
print("The residual for 4th degree fitting is", fit4[1])
print("The residual for 5th degree fitting is", fit5[1])
print("The residual for 6th degree fitting is", fit6[1])
```
