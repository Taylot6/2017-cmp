```python
#Trevor Taylor

#First we must download the iris flower data

from sklearn.datasets import load_iris
dataset = load_iris()

#We must create a data set and target set to see how well we can predicted a target based on data

X = dataset.data
Y = dataset.target
print(X)
print(Y)
```

```python
#It is much easier to work with 1 and 0s on a computer instead of normal values
#To do this we take each of the columns and find the adverage of each
#We assign a 1 to values over the adverage and a 0 to those bellow

#To do this I split the data into 4 seperate arrays since I was having trouble doing this otherwise

import numpy as np
X_d0 = np.array( X[:,0] > X[:,0].mean(), dtype = 'int')
X_d1 = np.array( X[:,1] > X[:,1].mean(), dtype = 'int')
X_d2 = np.array( X[:,2] > X[:,2].mean(), dtype = 'int')
X_d3 = np.array( X[:,3] > X[:,3].mean(), dtype = 'int')

#X_d = np.array( X > X.mean(), dtype='int')
```

```python
#Now since we have four arrays and we want one we must join them together
#In order to do this they must all have the same shape so we use resape to change them from a (150,) to (150,1)

X_d0r = X_d0.reshape(150,1)
X_d1r = X_d1.reshape(150,1)
X_d2r = X_d2.reshape(150,1)
X_d3r = X_d3.reshape(150,1)
```

```python
from collections import OrderedDict
import pandas as pd

#Using pandas to look at the data

Iris_data_d = OrderedDict({'Sepal Length ': X_d0r[:,0],
             'Sepal Width': X_d1r[:,0],
             'Petal Length': X_d2r[:,0],
             'Petal Width': X_d3r[:,0],
             'Label': Y})

df = pd.DataFrame(Iris_data_d)
df
```

```python
#Also we must join the target set to the other set and to do so we need it to change its dimentions as well

Y_d = Y.reshape(150,1)
```

```python
#Using np.hstack we can rejoin the arrays to form one full array

Z_d = np.hstack([X_d0r,X_d1r,X_d2r,X_d3r,Y_d])
```

```python
#To get the one rule algorythm we must see which one of our four variables correlates to the flower type the most
#What I did here was test how often a variable was over its average and how often that correlated to a certain flower type
#i = 0 means Speal length
#i = 1 means Speal width
#i = 2 means Petal length
#i = 3 means Petal width
#j = 0 means Setosa
#j = 1 means Versicolor
#j = 2 means Virginica

values = []
for i in range(4):
    for j in range(3):
        selection = np.logical_and(Z_d[:,i]>.9,Z_d[:,4] == j)
        values.append(('i', i, 'j', j,'value', len(Z_d[selection]), 'percentage is', ((len(Z_d[selection])/len(Z_d[:,i]))*100)))
```
