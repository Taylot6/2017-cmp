```python
#Trevor Taylor

import numpy as np
from scipy.integrate import quad
from math import exp
import matplotlib.pyplot as plt

#Constants
V = 1000 * (10**-6)     #m**3
p = 6.022 * (10 ** 28)   #Kg / m**3
k_b = 1.381 * (10**-23) #J / K
theta = 428             #K

#Function for integral
f = lambda x:x**4*exp(x)/(exp(x)-1)**2

#Function called heat cap to calulate heat capacity based on Debye's theory
def Heat_cap(T):
    return 9 * V * k_b * p * ((T/theta)**3) * quad(f, 0, theta/T)[0]

#Array and range created to create graph
Heat_array = []
N_series = range(5,500,1)
             
#For loop to calculate the points for the graph and to put them into an array
for T in N_series:
    Heat_array.append(Heat_cap(float(T)))
             
plt.plot(N_series,Heat_array)
plt.xlabel("Temperature(K)")
plt.ylabel("$C_V$ (J/K)")
plt.title("Heat Capacity vs Temperature")
plt.show()
```
