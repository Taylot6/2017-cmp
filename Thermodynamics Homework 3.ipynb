```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from numba import jit

def Multiplicity(N):
    
    N_series = range(N+1)
    empty = []
    
    for n in range(N+1):
        M = factorial(N)/(factorial(n) * factorial(N-n))
        empty.append(M)
    
    plt.bar(N_series, empty)
    plt.title('%s Coints Flipped' %N)
    plt.xlabel('Number of Heads')
    plt.ylabel('Multiplicity of the number of heads')
    
    plt.show()
        
```

```python
Multiplicity(10)
Multiplicity(15)
Multiplicity(30)
Multiplicity(100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def Multiplicity(N,q):
    
    N_series = range(q+1)
    empty = []
    
    for q in range(q+1):
        M = factorial(q+N-1)/(factorial(q)*factorial(N-1))
        empty.append(M)
    
    plt.bar(N_series, empty)
    plt.title('Einstein model with %s atoms and %s energy' %(N,q))
    plt.xlabel('q')
    plt.ylabel('Multiplicity of q')
    
    plt.show()
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def M_a(N_a,q):
    return factorial(q+N_a-1)/(factorial(q)*factorial(N_a-1))
    
def M_b(N_b,q):
    return factorial((100-q)+N_b-1)/(factorial((100-q))*factorial(N_b-1))

def Multiplicity(N_a,N_b,q):
    
    N_series = range(q+1)
    empty = []
    
    for q in range(q+1):
        M1 = factorial(q+N_a-1)/(factorial(q)*factorial(N_a-1))
        M2 = factorial((100-q)+N_b-1)/(factorial((100-q))*factorial(N_b-1))
        
        Mt = M1 * M2
        
        empty.append(Mt)
    
    plt.bar(N_series, empty, width = 1)
    plt.title('Solids with %s and %s atoms with %s energy' %(N_a,N_b,q))
    plt.xlabel('q')
    plt.ylabel('Multiplicity of q')
    
    plt.show()
```
