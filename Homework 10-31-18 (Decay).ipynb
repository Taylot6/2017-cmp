```python
#Trevor Taylor

%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

def Decay(N_atom,N_time):
    tau = 3.053*60
    mu = np.log(2)/tau
    N_Pb = np.zeros([N_time]) 
    N_Tl = N_atom * np.ones([N_time])  
    for i in range(N_atom):    
        r = int(-np.log(1-(np.random.random()))/mu) 
        if r < N_time:     
            N_Pb[r:] += 1  
            N_Tl[r:] -= 1 
        
    plt.plot(N_Tl, label='Tl')
    plt.plot(N_Pb, label='Pb')
    plt.title("Decay")
    plt.xlabel("Time (s)")
    plt.ylabel("Number of atoms")

    plt.legend()
    plt.show()
```

```python
Decay(1000,1000)
```
